```python
#Due to unnecessary HTML tags it is difficult to describe the textual content
from bs4 import BeautifulSoup
```

```python
def strip_html_tags(text):
    soup = BeautifulSoup (text,"html.parser")
    [s.extract() for s in soup (['iframe','script'])]
    stripped_text = soup.get_text()
    stripped_text=re.sub(r'[\r|\n|\r\n)]+', '\n', stripped_text)
    return stripped_text
```

```python
clean_content = strip_html_tags(text)
print(clean_content)
```

```python
# Import 5 Best Oscar Speeches and Tokenize them
#https://www.usatoday.com/story/entertainment/movies/oscars/2020/02/02/best-oscar-acceptance-speeches-ever-sally-field-halle-berry/4594790002/
from bs4 import BeautifulSoup
contents = urllib.request.urlopen ('https://www.usatoday.com/story/entertainment/movies/oscars/2020/02/02/best-oscar-acceptance-speeches-ever-sally-field-halle-berry/4594790002/').read()    
```

```python
speechtext=[]
heading=[]
```

```python
soup = BeautifulSoup(contents, 'lxml')
tags = soup.find_all(['p'])
for tag in tags:
    speechtext.append(tag)
    print(" ".join(tag.text.split()))

heads = soup.find_all(['h2'])
for head in heads:
    heading.append(head)
    print(" ".join(head.text.split()))
```
